## 次元削減

- 主成分分析

```python
import numpy as np
import matplotlib.pyplot as plt
# シード値を固定
np.random.seed(123)
# 0以上１未満の一様乱数を５０個生成
X = np.random.random(size=50)
# Xを２倍した後に、０以上１未満の一様乱数を０.５倍して足し合わせる
Y = 2*X + 0.5*np.random.rand(50)
# 散布図をプロット
fig, ax = plt.subplots()
ax.scatter(X, Y)
plt.show()
```

```python
from sklearn.decomposition import PCA
# 主成分のクラスをインスタンス化
pca = PCA(n_components=2)
# 主成分分析を実行
X_pca = pca.fit_transform(np.hstack((X[:, np.newaxis], Y[:, np.newaxis])))
```

```python
# 主成分分析の結果得られた座標を散布図にプロット
fig, ax = plt.subplots()
ax.scatter(X_pca[:, 0], X_pca[:, 1])
ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_xlim(-1.1, 1.1)
ax.set_ylim(-1.1, 1.1)
plt.show()
```
